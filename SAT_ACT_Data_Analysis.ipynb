In the data set the participation values are in percentage and they are processed as string. We need to make it the float data type.

```python
SAT17['Participation'] = SAT17['Participation'].str.rstrip('%').astype('float')/100
SAT18['Participation'] = SAT18['Participation'].str.rstrip('%').astype('float')/100
```

```python
pp.figure(figsize=(12, 10))
sns.set(font_scale = 2)
pp.title('Correlation of all data from SAT 2017', size=20)
sns.heatmap(SAT17.corr(), annot = True, cmap = 'viridis')
pp.xticks(rotation=40)
pp.show()
```

```python
pp.figure(figsize=(12, 10))
sns.set(font_scale = 2)
pp.title('Correlation of all data from SAT 2018', size=20)
sns.heatmap(SAT18.corr(), annot = True, cmap = 'viridis')
pp.xticks(rotation=40)
pp.show()
```

This can be seen from this plot that in both years 2017 and 2018 the total scores have declined against increase in participation.
